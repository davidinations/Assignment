```python
num = [-1, 0, 1, -2, -1, 2, -4]

def sumList(nums):
    result = []
    nums.sort()
    for i in range(len(nums)-2):
        if i > 0 and nums[i] == nums[i-1]:
            continue
        j = i+1
        k = len(nums)-1
        while j < k:
            total = nums[i] + nums[j] + nums[k]
            if total == 0:
                result.append([nums[i], nums[j], nums[k]])
                while j < k and nums[j] == nums[j+1]:
                    j += 1
                while j < k and nums[k] == nums[k-1]:
                    k -= 1
                j += 1
                k -= 1
            elif total < 0:
                j += 1
            else:
                k -= 1
    return result

listOfList = sumList(num)
result = ', '.join([str(lst) for lst in listOfList])
print(result)


```
