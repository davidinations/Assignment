```python
def modeValue(list1):
    if len(list1) == 0:
        return None
    else:
        if len(list1) == len(set(list1)):
            return None
        else:
            return max(set(list1), key = list1.count)

assert modeValue([]) == None
assert modeValue([1, 2, 3, 4, 4]) == 4
assert modeValue([1, 1, 2, 3, 4]) == 1
assert modeValue([1.5, 1.4, 2.1, 3.3, 4.9]) == None
```

```python
def checkBalance(text):
    """Counts the number of occurrences of each character in a string.

    Args:
        text: The string to count the characters of.

    Returns:
        A dictionary mapping each character to its number of occurrences.
    """
    dictionary = {}
    for character in text:
        if character in dictionary:
            dictionary[character] += 1  # it will add 1 to the value of the key if character already exists
        else:
            dictionary[character] = 1   # it will add the character as a key and 1 as its value if character does not exist
    values = set(dictionary.values())   # it will create a set of values of the dictionary so that we can check if all the values are same or not
    if len(values) == 1:
        return 'Yes'
    else:
        return 'No'

assert checkBalance('aa') == 'Yes'
assert checkBalance('aabb') == 'Yes'
assert checkBalance('aabbbccc') == 'No'
assert checkBalance('aabbbcccc') == 'No'
```
