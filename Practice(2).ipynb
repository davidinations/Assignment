```python
def isLeapYear(year):
    """
    _Menebak apakah tahun kabisat atau tidak_

    Args:
        year (int): tahun yang ingin diperiksa

    Returns:
        Boolean: True jika tahun kabisat, False jika tidak
    """    
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

def isValidDate(year, month, day):
    """
    _Menebak apakah tanggal valid atau tidak_

    Args:
        year (int): tahun yang ingin diperiksa
        month (int): bulan yang ingin diperiksa
        day (int): tanggal yang ingin diperiksa

    Returns:
        Boolean : True jika tanggal valid, False jika tidak
    """    
    if month < 1 or month > 12:
        return False
    if day < 1 or day > 31:
        return False
    if month == 2:
        if isLeapYear(year):
            if day > 29:
                return False
        else:
            if day > 28:
                return False
    if month == 4 or month == 6 or month == 9 or month == 11:
        if day > 30:
            return False
    return True

assert isValidDate(2001, 2, 29) == False
assert isValidDate(2029, 13, 1) == False
assert isValidDate(1000000, 1, 1) == True
assert isValidDate(2015, 4, 31) == False
```

```python
def commaFormat(number):
  """
  _Mengubah angka menjadi format yang lebih mudah dibaca_
  Args:
      number (integer or float): angka yang ingin diubah

  Returns:
      string: angka yang sudah diubah menjadi format yang lebih mudah dibaca
  """  
    if isinstance(number, float):
      return commaFormat(int(number)) + '.' + str(number).split('.')[1]
    else:
      if number < 0:
        return '-' + commaFormat(-number)
      if number < 1000:
        return str(number)
      else:
        return commaFormat(number // 1000) + ',' + str(number % 1000).zfill(3)

assert commaFormat(1) == '1'
assert commaFormat(10) == '10'
assert commaFormat(100) == '100'
assert commaFormat(1000) == '1,000'
assert commaFormat(10000) == '10,000'
assert commaFormat(100000) == '100,000'
assert commaFormat(1000000) == '1,000,000'
assert commaFormat(1234567890) == '1,234,567,890'
assert commaFormat(1000.123456) == '1,000.123456'
```
